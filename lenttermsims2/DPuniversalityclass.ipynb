```python
import numpy as np
Bc = 0.18547368

modelb = (np.abs(b - Bc))**0.5
modelv = (np.abs(b - Bc))**(-1) 
modelv[modelv > 100] = 200

modelc = (np.abs(b - Bc))**(-1.5) 
modelc[modelc > 300] = 400
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Spreading Rate')
ax1.set_ylabel('Row Count', color=color)
ax1.plot(b, s1starav, color=color, marker = 'o', label= 'S1* Average (Order Parameter)')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:blue'
color = 'tab:blue'
ax2.plot(b, modelb, color=color, marker = 'o', label= 'IMF model of Order Parameter')
ax1.set_ylabel('', color=color)


fig.legend(loc = 'lower right')
plt.title('Spatial Plot with Spreading Rate')
fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```
